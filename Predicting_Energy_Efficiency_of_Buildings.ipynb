```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Load the dataset into pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/Hamoye/energydata_complete.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.describe()
```

```python
# check for missing records
for col in df.columns:
  missing_values = df[col].isnull().sum()
  total = df[col].count()
  perct_missing = missing_values/total*100
  print (f"{col} has {missing_values},{ perct_missing}% ")
```

```python
df.dtypes
```

```python
x_train = df[['T2']]
y_train = df[['T6']]
```

```python
from sklearn.linear_model import LinearRegression
linear_model = LinearRegression()
# Fit the model to the training dataset
linear_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predicted = linear_model.predict(x_train)

rmse = np.sqrt(mean_squared_error(y_train, predicted))

rmse
```

```python
df.drop(['date', 'lights'], axis=1, inplace = True)
```

```python
y = df['Appliances']
x = df.drop('Appliances', axis = 1)

# Now we split our dataset into the training and testing dataset. Recal that we had earlier segmented features and tarhet variables
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,  y, test_size=0.3, random_state=42)
```

```python
# Firstly, we normalize our dataset to a common scale using the min max scaler
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train_sc = pd.DataFrame(scaler.fit_transform(x_train))
```

```python
x_test_sc= pd.DataFrame(scaler.transform(x_test))
```

```python
linear_model = LinearRegression()
# Fit the model to the training dataset
linear_model.fit(x_train_sc, y_train)
# Obtain predictions
predicted_values = linear_model.predict(x_train_sc)
```

```python
# MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_train, predicted_values)
round(mae, 3)
```

```python
rmse = np.sqrt(mean_squared_error(y_train, predicted_values))
round(rmse, 3)
```

```python
predicted_values = linear_model.predict(x_test_sc)
```

```python
# MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
# Ridge regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge()
ridge_reg.fit(x_train_sc, y_train)
```

```python
predict_ridge_values = ridge_reg.predict(x_test_sc)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predict_ridge_values))
round(rmse, 3)
```

```python
# Lasso regression
from sklearn.linear_model import Lasso
Lasso_reg = Lasso()
Lasso_reg.fit(x_train_sc, y_train)
```

```python
predict_lasso_values = Lasso_reg.predict(x_test_sc)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predict_lasso_values))
round(rmse, 3)
```

```python
weights = pd.Series(Lasso_reg.coef_, x_train.columns).sort_values()
weights_df = pd.DataFrame(weights).reset_index()
weights_df.columns = ['Features', 'Weight']
```
